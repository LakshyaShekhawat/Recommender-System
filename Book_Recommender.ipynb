#Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num_ratings'}, inplace = True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg_ratings'}, inplace = True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popularity_df
```

```python
popular_df = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_ratings', ascending = False).head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

#Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
padhe_likhe_users = x[x].index
```

```python
padhe_likhe_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title')['Book-Rating'].count() >= 50
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
similarity_score_df = pd.DataFrame(similarity_score)
similarity_score_df
```

```python
def recommend(book_name):
  #index fetch
  index = np.where(pt.index == book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True)[1:5]

  data = []
  for i in similar_items:
    item = []
    temp_df = books[books['Book-Title'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

    data.append(item)

  return data
```

```python
recommend('The Fellowship of the Ring (The Lord of the Rings, Part 1)')
```
